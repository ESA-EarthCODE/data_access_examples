```python
import contextily as cx
import matplotlib.pyplot as plt
import rioxarray

# select an url from the catalog or the earthcode STAC API
cog_url = "https://s3.waw4-1.cloudferro.com/EarthCODE/OSCAssets/eo4flood/Flood_extent/Brahmaputra/Flood_extent/From_VIIRS/Brahamaputra_20241010/EO4FLOOD_VIIRS_FloodedExtent_30m_braa_20241011_20241014_20250728.tif"

rds = rioxarray.open_rasterio(cog_url, masked=True)
rds_3857 = rds.rio.reproject("EPSG:3857")

fig, ax = plt.subplots(figsize=(12, 10))
rds_3857.squeeze().plot(ax=ax, cmap="Blues", alpha=0.6, add_colorbar=True)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron)
```
