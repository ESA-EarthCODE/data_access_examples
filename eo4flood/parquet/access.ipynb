```python
import pandas as pd
import geopandas as gpd
import contextily as cx
import matplotlib.pyplot as plt
```

```python
# select an url from the catalog or the earthcode STAC API
url = 'https://s3.waw4-1.cloudferro.com/EarthCODE/OSCAssets/eo4flood/width_series/Congo_width_timeseries.parquet'
```

```python
df = pd.read_parquet(url)
# create geodataframe for easier plotting
gdf = gpd.GeoDataFrame(df, 
                 geometry=gpd.points_from_xy(df['gauge_lon'], df['gauge_lat'], crs='epsg:4326')
                 )
```

```python


# 1. Filter the width data and reproject it to Web Mercator (EPSG:3857)
gdf_filtered = gdf[~gdf.w.isna()].to_crs(epsg=3857)
ax = gdf_filtered.plot(figsize=(10, 10), alpha=0.7, edgecolor='black')
cx.add_basemap(ax)
ax.set_axis_off()
```
